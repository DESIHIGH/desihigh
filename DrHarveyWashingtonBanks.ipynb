### A formal description of clustering

Cosmological _'perturbation theory'_ is a tool we use to describe the relationship between the 'clustering' (relative numbers amongst different large volumes) of _all_ matter and galaxies.  The equation relating these quantities may look confusing at first glance, but it's really just a simple description of this picture, written as a polynomial with (perhaps more than one) independent variable! 

Perhaps you have experience working with 'polynomials' that have one 'independent variable', e.g. the equation of a straight line $f(x)= mx + b$, where x is our polynomial of one input variable, e.g. the number of customers served, $f(z)$, for a given number of staff, $x$.  Similarly, the equation of a quadratic function $f(x) = ax^2 + bx + c$, might describe more complex behaviour for the same situation. 

Others may have worked with equations possessing two independent variables such as the equation of a circle: $x^2+y^2=r^2$.  Here $(x,y)$ describes a position in the plane, $r$ is the radius of a circle, and points in the plane that satisfy this _constraint_ would draw a circle.

If you're used to seeing these equations, the equation for clustering won't look too bad!  Otherwise, you'll soon get used to it.  

Without further ado, I present one of the many equations I use to describe **galaxy clustering**:

$\delta_g = b_1\delta_m + \frac{b_2}{2}\left[\delta^2_m-\langle \delta^2_m \rangle\right] +\frac{b_s}{2}\left[s^2-\langle s^2 \rangle\right] + \ldots$

Let's take a minute to talk about what each piece of this means, starting with our first piece: 

$\delta_g = b_1\delta_m + \ldots$

The Greek letter delta ($\delta$) represents an _overdensity_ - a measure of how the relative density of the volume you're looking at compares to the average density of the Universe.  The subscript $g$ describes this for galaxies and $m$ does the same for total matter.  Let's take an example, say our avg. number of galaxies in a volume can be represented by $\bar n_g$, a volume with $\delta_g = \left (n / \bar n\right ) - 1 =2$, would have $\times 3$ the number of galaxies than normal, by definition!  While $b_1$ is an unknown number (or _'bias coefficient'_) that allows sufficient flexibility to describe the Universe we see.  For our example, if $\delta_g = 2$ and $b_1=4$, then we have $\delta_m = 0.5$, or the same volume with $\times 1.5$ the amount of _matter_ than average, despite a much greater number of _galaxies_.  Galaxy formation is complex, so we simply write a flexible model for how galaxies form in the Universe and find the one (set of bias coefficients) that describes our data the best!

Let's take a toy one-dimensional model of the Universe where Mario, or you!, can only walk in a straight line:

```python
# Distance from the starting point, right is positive, left is negative.  
displacements = np.arange(-10, 10., 1)
```

Where the fluctuations in the matter density from point to point are well described by a normal distribution, 

```python
delta_m = np.random.normal(loc=0.0, scale=1.0, size=len(displacements))
```

```python
pl.plot(displacements, delta_m, c='k', lw=0.2)

pl.xlabel('Distance from the start.')
pl.ylabel('$\delta_m$')
```

We see that, far to our left, $\delta_m = -1.5$, which represents a large _void_, with relatively very little mass.  While to our right, the large spike would be a dense cluster of mass in our toy Universe.

What can our first simple model for galaxy formation describe?  Let's take a range for the values $b_1$ _might_ be,   

```python
b1s = np.arange(0.5, 2.5, 0.5)
b1s = b1s[b1s != 1.0]
```

```python
pl.plot(displacements, delta_m, c='k', lw=0.5)

for b1 in b1s:
    pl.plot(displacements, b1 * delta_m, lw=0.2, label='$b_1: {:.1f}$'.format(b1))

pl.xlabel('Distance from the start.')
pl.ylabel('$\delta_m$')
pl.legend(frameon=False)
```

As advertised, we have a range of Universes in which galaxies form differently!  Each color shows such a Universe, where the number of galaxies varies, despite the underlying amount of matter being the same between them.

Like our toy model for a coffee shop, a more developed model 

$\delta_g = b_1\delta_m + \frac{b_2}{2}\left[\delta^2_m-\langle \delta^2_m \rangle\right]$ 

describes more complex behaviour between these same relative densities.  In our first model, ($\delta_g / \delta_m$) is always the constant $b_1$, whereas this is relaxed in our second model.  Often, this is adopted to better describe more extreme volumes in the Universe, in a particular dense region of the Universe when $\delta_m$ is a large number, we would see very different behaviour for our first 'linear' model.  Let's see an example, 

```python
b2 = 1.5
```

```python
pl.plot(displacements, delta_m, c='k', lw=0.5)

for b1 in b1s:
    pl.plot(displacements, b1 * delta_m + 0.5 * b2 * (delta_m**2. - np.mean(delta_m**2.)), lw=0.2, label='$b_1: {:.1f}$'.format(b1))

pl.xlabel('Distance from the start.')
pl.ylabel('$\delta_m$')
pl.legend(frameon=False)
```

For volumes close to the average density $\delta_m \simeq 0$ we see very little difference from the first model, whereas the number of galaxies in our dense cluster has doubled!  We haven't seen $\langle \delta_m \rangle$ before, but it's a simple idea:  this represents the average $\delta_m$ over all volumes.  As we have defined this to be relative fluctuations it follows $\langle \delta_m \rangle = \langle n \rangle / \bar{n} - 1$, but $\langle n \rangle$ is our avg. density $\bar n$, so $\langle \delta_m \rangle$ must be zero.  Applying the same idea to $\langle \delta_m \rangle$ means our model must contain this strange $\langle \delta_m^2 \rangle$ term, or be unphysical!        

Our final, and most advanced, set of model Universes are described by:

$\delta_g = b_1\delta_m + \frac{b_2}{2}\left[\delta^2_m-\langle \delta^2_m \rangle\right] +\frac{b_s}{2}\left[s^2-\langle s^2 \rangle\right] + \ldots$

In this case, the number of galaxies in a given volume isn't determined solely from the amount of matter, $\delta_m$.  Rather, it may also be influenced by the _tidal force_ on that volume.  For instance, the tidal force on a proto-galaxy trying to form may be a 'differential shear' - a fancy term to describe an obvious idea that the galaxy may be feeling a stronger force pulling in one direction on one side of it, rather than the other.  This might cause the galaxy to be more, or less, likely to form.  We don't know, but it's possible!  We can then include this idea but adding the innocent looking $s$, a tidal 'tensor', which describes just this gravitational acceleration of two neighboring pieces of matter.  

While purely academic, even the professionals use these same exercises to better understand the underlying ideas and check their results (we do not condone any references to drug use in this paper!).

###  How complex does our model Universe need to be?

The only quantity we can directly measure and compare to our model is $\delta_g$.  How then do we know what different effects should appear on the right-hand side of the equation?  A full description requires answering two distinct questions: What are the properties, or 'fields' - a physical quantity such as temperature defined at every point in space, we might need to care about?  In the example above, these would be $\delta_m$ and $s$, but are there others?  In what ways can those fields be included and still be physical?  We had $\delta_m$ without problem, but we had to limit ourselves to $\delta_m^2 - \langle \delta_m^2 \rangle$ to be a valid description of nature.  

A fundamental principle of cosmological perturbation theory (today, you might design a better version!) is that gravity is the only long-range force we care about. This means that every field we include must describe gravity in one form or another.  One type of field that makes sense to include is the matter overdensity. We could treat normal matter and dark matter with different fields, but it's more common to put them together. Another type of field we could include is related to changes in the velocity of our matter since gravitational acceleration causes those changes.  We don't include a dependence of galaxy formation on the velocity of the dark matter at that point because of Special Relativity - the foundation of which is that the laws of nature are the same no matter what speed you're travelling at.  Other fields exist and could be included, but we'll stop here!

Once we have the fields we care about we can think about how the best ways to include them to make a more realistic model Universe.  **Symmetries** give us the answer to this question.  Try this: Take a piece of paper and draw a square.  Now rotate that square by 90 degrees. Notice how the square looks the same as it did before you rotated it. We call this rotation a symmetry of a square as it appears similarly after this rotation.  If instead you now rotate the square by some other angle that isn't a multiple of 90 degrees, you'll see that the square looks different. Those types of rotations are not symmetries of the square.

M. C. Escher has a lot of fun and success designing art that satisfied different symmetry requirements!  
What symmetries does this Escher work possess?  What if we printed it in black and white?  Escher's work
shares a lot of philosophy with Eisnstein's General Relativity, but we won't cover that here.

When we think about how to build up our clustering equation, or model Universe, we need to ensure that anything we add has the right symmetries of the problem at hand, otherwise it's not a valide description.  Both Copernicus and Einstein believed that physics understood everyone to be equal, and our model should satisfy their viewpoint also.  By imposing symmetries on our models, we are just stipulating that the laws of the Universe should be the same in one place, or at one speed, as at any other.  

In this example, this is why our model includes the square of $s^2$, but not $s$; If we were to include $s$, not all observers would agree.
